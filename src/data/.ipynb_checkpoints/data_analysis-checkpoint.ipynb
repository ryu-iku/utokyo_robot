```python
import numpy as np
import matplotlib.pyplot as plt
import math

from math import pi
```

```python
def si_to_ori(si):
    return ((1. - 2. * si / 40.) * pi) % (2. * pi)
```

```python
def moving_average(arr, mean_window):
    mean_ranges = []
    sample_n = len(arr)
    for i in range(sample_n):
        s_sum = 0
        for j in range(mean_window):
            cal_i = (i - int(mean_window / 2) + j) % sample_n
            s_sum += arr[cal_i]
        mean_ranges.append(s_sum / mean_window)
    return mean_ranges
```

```python
sample_n = 41
shortest_road = 1.5

quarter_lsi = int((sample_n - 1) / 4)
back_i = 0
right_i = quarter_lsi
front_i = quarter_lsi * 2
left_i = quarter_lsi * 3
right_front_i = int((sample_n - 1) * 1 / 8)
right_back_i = int((sample_n - 1) * 3 / 8)
left_back_i = int((sample_n - 1) * 5 / 8)
left_front_i = int((sample_n - 1) * 7 / 8)

def update_sensor_data_analysis(arr):
#     for i in range(len(arr)):
#         print(i, arr[i])
    arr = moving_average(arr, int(sample_n / 8))
#     arr = moving_average(arr, 3)
#     for i in range(len(arr)):
#         print(i, arr[i])

    max_sensor_arr = []
    msa_front = []

    for i in range(sample_n):
        curr_s = arr[i]
        prev_s = arr[i - 1]
        next_s = arr[(i + 1) % sample_n]

        if curr_s > prev_s and curr_s >= next_s:
            max_sensor_arr.append(i)
            if curr_s > shortest_road and i >= right_back_i and i <= left_back_i:
                msa_front.append(i)
                
    return max_sensor_arr, msa_front
```

```python
data_files = ["183625.record_start","221329.record","221437.record_corner","221551.record_corner",
             "221739.record","221812.record_corner"]
dt = np.genfromtxt(data_files[0], delimiter=",")
```
